```python
# Transfer learning (MobileNetV2) вҖ” СҖРөРәРҫРјРөРҪРҙСғРөСӮСҒСҸ Р·Р°РҝСғСҒРәР°СӮСҢ РІ Google Colab
# РЈРұРөРҙРёСҒСҢ, СҮСӮРҫ Р°СҖС…РёРІ СҖР°СҒРҝР°РәРҫРІР°РҪ РІ train_path (РҪР°РҝСҖРёРјРөСҖ /content/citys)

import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns

#!unzip 'drive/MyDrive/РқРөР№СҖРҫСҒРөСӮРё/citys.zip'

# РҹР°СҖР°РјРөСӮСҖСӢ вҖ” РјРҫР¶РҪРҫ РјРөРҪСҸСӮСҢ
train_path = '/content/citys/citys'     # РҝСғСӮСҢ Рә СҖР°СҒРҝР°РәРҫРІР°РҪРҪРҫР№ РҝР°РҝРәРө СҒ РҝР°РҝРәР°РјРё-РәР»Р°СҒСҒР°РјРё
img_height, img_width = 320, 320  # СғРІРөР»РёСҮРёСӮСҢ СҖР°Р·СҖРөСҲРөРҪРёРө РҝРҫРјРҫРіР°РөСӮ РҝСҖРөРҙРҫРұСғСҮРөРҪРҪСӢРј РјРҫРҙРөР»СҸРј
batch_size = 32
seed = 42

# Р“РөРҪРөСҖР°СӮРҫСҖСӢ (РёСҒРҝРҫР»СҢР·СғРөРј preprocess_input MobileNetV2)
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input

train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=seed
)

val_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=False,
    seed=seed
)

num_classes = len(train_generator.class_indices)
class_names = list(train_generator.class_indices.keys())
print("Classes:", class_names)

# РЎРұРҫСҖРәР° РјРҫРҙРөР»Рё
base = MobileNetV2(include_top=False, weights='imagenet', input_shape=(img_height, img_width, 3))
base.trainable = False  # СҒРҪР°СҮР°Р»Р° СӮРҫР»СҢРәРҫ РіРҫР»РҫРІР°
inp = Input(shape=(img_height, img_width, 3))
x = base(inp, training=False)
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x)
out = Dense(num_classes, activation='softmax')(x)
model = Model(inp, out)

model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Callbacks
ckpt_path = '/content/best_mobilenetv2_pusher.h5'
callbacks = [
    ModelCheckpoint(ckpt_path, save_best_only=True, monitor='val_accuracy', mode='max'),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True, verbose=1)
]

# Train head
epochs_head = 20
history1 = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs_head,
    callbacks=callbacks,
    verbose=1
)

# Fine-tune: СҖР°Р·РјРҫСҖРҫР·РёРј СҮР°СҒСӮСҢ РұР°Р·РҫРІРҫР№ РјРҫРҙРөР»Рё
base.trainable = True
# СҖР°Р·РјРҫСҖРҫР·РёРј РҝРҫСҒР»РөРҙРҪРёРө N РұР»РҫРәРҫРІ (РҪР°СҒСӮСҖРҫР№РәР°)
fine_tune_at = 100  # РјРҫР¶РҪРҫ РҝРҫРҙРҝСҖР°РІРёСӮСҢ РІ Р·Р°РІРёСҒРёРјРҫСҒСӮРё РҫСӮ Р°СҖС…РёСӮРөРәСӮСғСҖСӢ
for layer in base.layers[:fine_tune_at]:
    layer.trainable = False
for layer in base.layers[fine_tune_at:]:
    layer.trainable = True

model.compile(optimizer=Adam(learning_rate=5e-5), loss='categorical_crossentropy', metrics=['accuracy'])
epochs_finetune = 40
history2 = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs_finetune,
    callbacks=callbacks,
    verbose=1
)

# РһСҶРөРҪРәР° РҪР° РІР°Р»РёРҙР°СҶРёРё: СҒРҫРұРөСҖС‘Рј РҝСҖРөРҙСҒРәР°Р·Р°РҪРёСҸ Рё РҫСӮСҮС‘СӮ
import numpy as np
val_steps = int(np.ceil(val_generator.samples / val_generator.batch_size))
y_pred_proba = model.predict(val_generator, steps=val_steps, verbose=1)
y_pred = np.argmax(y_pred_proba, axis=1)
y_true = val_generator.classes[:len(y_pred)]

print("Overall accuracy:", accuracy_score(y_true, y_pred))
print("Classification report:")
print(classification_report(y_true, y_pred, target_names=class_names, zero_division=0))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion matrix')
plt.show()

# Per-class accuracy
per_class = cm.diagonal().astype(float) / cm.sum(axis=1)
for i, cname in enumerate(class_names):
    print(f"Accuracy of {cname:12s}: {100*per_class[i]:.2f}%")
```
